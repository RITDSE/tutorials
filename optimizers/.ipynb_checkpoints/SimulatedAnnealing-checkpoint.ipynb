```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def likelihood(x):
    y = abs(.5*x) + np.sin(x)
    return y
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,likelihood(x));
```

```python
points = np.random.uniform(-10,10, size = 10)

plt.plot(x,likelihood(x));
plt.scatter(points,likelihood(points), color = "orange")
```

```python
temperature = 1

newPoints = points + temperature * np.random.normal(size=10)

for i in range(0,len(points)):
    if likelihood(newPoints[i]) < likelihood(points[i]):
        points[i] = newPoints[i]

```

```python
plt.plot(x,likelihood(x));
plt.scatter(points,likelihood(points), color = "orange")
```

```python
points = np.array([-10,-9,-8,-7,-6,5,4,7,8,9,10])
temperature = 10

mins = []
temps = []

for n in range(0,50):
    temperature = temperature/(n+1)

    newPoints = points + temperature * np.random.normal(size=len(points))

    for i in range(0,len(points)):
        if likelihood(newPoints[i]) < likelihood(points[i]):
            points[i] = newPoints[i]
            
    mins.append(min(likelihood(points)))
    temps.append(temperature)
    print(mins[n])
```

```python
plt.plot(x,likelihood(x));
plt.scatter(points,likelihood(points), color = "orange")
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,2,1)
steps = np.arange(1,len(temps)+1,1)
plt.plot(steps,mins)

plt.subplot(1,2,2)
plt.plot(steps,temps)
```
