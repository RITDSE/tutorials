# Matplotlib Tutorial

```python
#Remeber the .pyplot!
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic plots using pyplot

Matplotlib includes many useful plot types, as well as bells and whistles to customize all of them to fit your data visualization needs. 

https://matplotlib.org/stable/plot_types/index.html


### Line Graph

```python
y = [1,3,7,10]
x = [1,2,3,4]

plt.plot(x,y)
plt.title("X vs. Y", fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12)
plt.show()
```

### Changing Figure Size

```python
plt.figure(figsize=(12, 6), dpi=80)
plt.plot([1, 3, 7, 10], [1, 4, 9, 16])
plt.title("X vs. Y", fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12)
plt.show()
## Use shift + O to see complete ouputs on jupyter when you dont want the output to be scrollable
```

### Scatter Plot

```python
plt.figure(figsize=(8, 8))
x = np.random.random(100)
y = np.random.random(100)

plt.scatter(x,y, marker = "o", color = "b")
plt.title("X, Y Scatter", fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12);
```

### Bar Chart

```python
label = [1,2,3,4]
height = [10,5,3,25]

plt.figure(figsize = (12,6))
plt.title("Bar Chart", fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12)

plt.bar(label,height);
```

```python
#Bar Chart with Labels
label = [1,2,3,4]
height = [10,5,3,25]

plt.figure(figsize = (12,6))
plt.title("Bar Chart with Labels", fontsize = 12)
plt.xlabel('Person', fontsize = 12)
plt.ylabel("Projected Wealth (Millions)", fontsize = 12)

#Specify labels for every x tick
plt.xticks(ticks = label,labels=["Le","Tolu", "Karan", "Sujan"])

plt.bar(label,height);
```

### Histogram

```python
x = np.random.normal(size = 1000)

plt.figure(figsize = (12,6))
plt.title("Histogram", fontsize = 12)
plt.xlabel('X', fontsize = 12)
plt.ylabel("Count", fontsize = 12)

plt.hist(x);
```

### Mulitple Plots

Sometimes it is useful to put multiple plots on the same plot for comparison, this is easily achieveable in matplotlib

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.figure(figsize=(8,6))
plt.plot(t, t, 'r--', label = "1")
plt.plot(t, t**2, 'bs', label = "2")
plt.plot(t, t**3, 'g^', label = "3")

plt.title("Multiple X vs. Y", fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12)

plt.legend();
```

```python
x = np.random.normal(size = 100)
y = np.random.normal(size = 100)

x2 = np.random.normal(size = 100) - 2
y2 = np.random.normal(size = 100) -2

plt.figure(figsize = (8,8))

plt.scatter(x,y, label = "class 1")
plt.scatter(x2,y2, label = "class 2")

plt.title("Multi Class Scatter")
plt.ylabel('Y', fontsize = 12)
plt.xlabel('X', fontsize = 12)

plt.legend();
```

# Subplots

Somtimes it is useful to put multiple graphs on the same plot, we do this with subplots. Subplots work by making a grid of plots to us where we specify where on the grid a plot should be with a row column coordinate.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html



### adding in parameters and axis restrictions

```python
plt.plot([1, 3, 7, 10], [1, 4, 9, 16], 'ro')
plt.axis([0, 13, 0, 20])
plt.show()
```

### Mutiple plots at once

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

### Can drop arguments if you dont want to define lines

```python
# d makes diamonds and g- forces a green colored line
plt.plot(t, t, 'rd', t, t**2, t, t**3, 'g-')
plt.show()
```

### Making a df from cereal and plotting its correlation matrix

```python
df = pd.read_csv("cereal.csv")
pear_corr=df.corr(method='pearson')
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(pear_corr.corr(), fignum=f.number,cmap ='viridis')
plt.xticks(range(len(pear_corr.columns)),pear_corr.columns, rotation=45) 
plt.yticks(range(len(pear_corr.columns)),pear_corr.columns) 
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
iris = pd.read_csv("iris.data")

fig, ax = plt.subplots(figsize=(15,8))
ax = iris.boxplot()
plt.title("Multi Boxplot using pandas", loc="left")
plt.show()
```

```python
groups = iris.groupby("class")
plt.figure(figsize=(12,8))
colors = {'Iris-setosa':'darkviolet', 'Iris-versicolor':'cornflowerblue', 'Iris-virginica':'orchid'}
for name, group in groups:
    plt.plot(group["sepal length"], group["sepal width"], marker="o", linestyle="", 
    label=name, ms=10, color=colors[name], alpha=.8)
    m, b = np.polyfit(group["sepal length"], group["sepal width"], 1)
    plt.plot(group["sepal length"], m*group["sepal length"]+b, linewidth=5, 
    color=colors[name], alpha=.5)
    
plt.legend()
plt.title('Iris Species classification')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```
